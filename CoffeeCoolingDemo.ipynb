```python
import numpy as np
import CoolingRoutine as cr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
times = np.linspace(0, 20, 201)
```

```python
temps_beforecooling = cr.cube_before_cooling(times, **cr.cooling_parameters)
```

```python
plt.plot(times, temps_beforecooling)
plt.xlabel('Time (min)')
plt.ylabel('Coffee Temperature (degrees Celcius)')
```

```python
temps_aftercooling = cr.cube_after_cooling(times, **cr.cooling_parameters)
```

```python
plt.plot(times, temps_aftercooling)
plt.xlabel('Time (min)')
plt.ylabel('Coffee Temperature (degrees Celcius)')
```

```python
temps_cont = cr.continuous_sugar_cooling(times, **cr.cooling_parameters)
```

```python
plt.plot(times, temps_cont)
plt.xlabel('Time (min)')
plt.ylabel('Coffee Temperature (degrees Celcius)')
```

```python
temps_n = cr.sugar_specifiedtime_cooling(times, [50, 100, 150], **cr.cooling_parameters)
```

```python
plt.plot(times, temps_n)
plt.xlabel('Time (min)')
plt.ylabel('Coffee Temperature (degrees Celcius)')
```

```python
plt.plot(times, temps_beforecooling, label='cube before cooling')
plt.plot(times, temps_aftercooling, label='cube after cooling')
plt.plot(times, temps_cont, label='continuous')
plt.plot(times, temps_n, label='3 times of dropping cube')
plt.xlabel('Time (min)')
plt.ylabel('Coffee Temperature (degrees Celcius)')
plt.legend()
```
